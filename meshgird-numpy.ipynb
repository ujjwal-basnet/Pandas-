```python
x = np.linspace(-1 , 1  , 10 )
y  = np.linspace(-1 , 1 , 10 )
```

```python
x == y  #both x and y are equal 
```

### you could write " sin(x) * cos(y) inorder to calculate sin(x  ,y )

```python
np.sin(x) * np.cos(y)
```

### but stil what you have got is in 1 -d array 

```python
(np.sin(x) * np.cos(y)).shape
```

## Thats where meshgid comes in 

```python
X  , Y  = np.meshgrid(x , y )
```

```python
X.shape 
```

```python
fig ,  ( ax1 , ax2 ) = plt.subplots(1 ,2 , figsize = (12 , 5 ) )
ax1.scatter(X , Y)
ax2.plot(X , Y)
plt.show()

#grid plot
```

calculating sin(x , y )

```python
np.sin(X) * np.cos(Y)
```

```python
#comprarisom side by side
fig , (ax1 , ax2 )  = plt.subplots(1 , 2  ,figsize = (12 , 5 ))
ax1.plot(np.sin(X) * np.cos(Y))
ax2.plot(np.sin(x) * np.cos(y))
plt.show()
```
